```python
from sklearn import preprocessing
label_encoder_1=preprocessing.LabelEncoder()
label_encoder_2=preprocessing.LabelEncoder()
label_encoder_3=preprocessing.LabelEncoder()
label_encoder_4=preprocessing.LabelEncoder()
label_encoder_5=preprocessing.LabelEncoder()
label_encoder_6=preprocessing.LabelEncoder()
label_encoder_7=preprocessing.LabelEncoder()
```

```python
def pipeline(dataset):
  dataset = dataset.drop(columns=['PassengerId','Name'])
  dataset['Pclass']=label_encoder_1.fit_transform(dataset['Pclass'])
  dataset['Sex']=label_encoder_2.fit_transform(dataset['Sex'])
  dataset['SibSp']=label_encoder_3.fit_transform(dataset['SibSp'])
  dataset['Parch']=label_encoder_4.fit_transform(dataset['Parch'])
  dataset['Ticket']=label_encoder_5.fit_transform(dataset['Ticket'])
  dataset['Cabin']=label_encoder_6.fit_transform(dataset['Cabin'])
  dataset['Embarked']=label_encoder_7.fit_transform(dataset['Embarked'])
  return dataset
```

```python

dataset = pipeline(df)
dataset
```

```python
X=dataset.drop(columns=['Survived'])
y=dataset['Survived']
X=X.to_numpy()
y=y.to_numpy()
print(X)
print(y)
```

```python
y=np.reshape(y,(-1,1))
y.shape
```

```python
def sigmoid(z):
  z=np.float128(z)
  return 1/(1+np.exp(-z))
```

```python
def accuracy(y,h):
  n=h.shape[0]
  acc=0
  for i in range(n):
    if((h[i]>=0.5 and y[i]==1) or (h[i]<0.5 and y[i]==0)):
      acc+=1
      #print(h[i],' ',y[i])
  print(f"Accuracy= {acc/n}\n")
```

```python
def logistic_regression(X,y,lr,num_iterations,lamda_reg):
  n,d=X.shape
  w =np.reshape(np.random.rand(d),(-1,1))
  for i in range(num_iterations):
    print(f"Iteration Number :{i+1}\n")
    y_cap=np.dot(X,w)
    h=sigmoid(y_cap)
    accuracy(y,h)
    gradient=(1/n)*(np.dot(X.T,h-y)+lamda_reg*w)
    w=w-lr*gradient
  return w

```

```python
lr = 0.000005
num_iterations = 5000
lambda_reg = 0.0001
w = logistic_regression(X,y,lr,num_iterations,lambda_reg)
```
